<h1> Checking equilibrium by looking at internal energies </h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_energy_file(fname):

    dat = pd.read_fwf('../%s/out/output/DNA-oxdna.log.energy.dat'%(fname,),header = None)
    dat.columns = ['t_steps','IE','KE','TE'] #time steps / internal energy / kinetic energy / total energy
    
    return dat
```

```python
energy_files = {}
fnames = ['%s_%s'%(i,j,) for i in ['Seeman','Turberfield'] for j in [0,2,4]]
```

```python
for fname in fnames:
    energy_files[fname] = read_energy_file(fname)
```

```python
color_files = {}
color_files['Seeman_0'] = (0.2,0,0)
color_files['Seeman_2'] = (0.5,0,0)
color_files['Seeman_4'] = (1.0,0,0)

color_files['Turberfield_0'] = (0,0,0.2)
color_files['Turberfield_2'] = (0,0,0.5)
color_files['Turberfield_4'] = (0,0,1.0)
```

```python
f,ax = plt.subplots(1,2,figsize = (6,4),sharey = True,sharex = True)

for sim in energy_files:
    if sim[0] == 'S':
        ax[0].plot(energy_files[sim]['t_steps'], 
            energy_files[sim]['IE'],
            color = color_files[sim],label = sim)
    elif sim[0] == 'T':
        ax[1].plot(energy_files[sim]['t_steps'], 
            energy_files[sim]['IE'],
            color = color_files[sim],label = sim)
        
for a in ax:
    a.legend()
        
ax[0].set_xlabel('time steps')
ax[0].set_ylabel('internal energy per nucleotide (oxDNA units)')
        
plt.tight_layout()

plt.savefig('TvsS_tetrahedron_internal_energies.pdf')
```

<p> All internal energies have approximately plateaued. Note that internal energies are initially more negative than at equilibrium as MRDNA places nucleotides in their most enthalpically favored states. </p>
