```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def get_energy(path):
    data = pd.read_fwf(path,
           header = None)
    return data[1]

fname = '../brady_tetrahedron/ssDNA0/out/output/DNA-oxdna.log.energy.dat'

fnames = [
    '../brady_tetrahedron/ssDNA0/out/output/DNA-oxdna.log.energy.dat',
    '../brady_tetrahedron/ssDNA2/out/output/DNA-oxdna.log.energy.dat',
    '../brady_tetrahedron/ssDNA4/out/output/DNA-oxdna.log.energy.dat',
    '../brady_tetrahedron_2/ssDNA0/out/output/DNA-oxdna.log.energy.dat',
    '../brady_tetrahedron_2/ssDNA2/out/output/DNA-oxdna.log.energy.dat',
    '../brady_tetrahedron_2/ssDNA4/out/output/DNA-oxdna.log.energy.dat'    
]

data = []

for i in fnames:
    data.append(get_energy(i))



f,ax = plt.subplots(1,2,figsize = (10,5),sharey = True)

ax[0].plot(data[0],'r--')
ax[0].plot(data[1],'g--')
ax[0].plot(data[2],'b--')

ax[1].plot(data[3],'r--')
ax[1].plot(data[4],'g--')
ax[1].plot(data[5],'b--')
```
